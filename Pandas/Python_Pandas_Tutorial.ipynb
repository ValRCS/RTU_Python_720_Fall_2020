## Two approaches IN PLACE and NOT IN PLACE when working with dataframe methods

### IN PLACE

IN PLACE will modify the dataframe directly. It is faster and uses less memory, but it is not recommended for production code, because it is harder to debug and understand. Still very convenient for quick and dirty data analysis.

````python
## We can renamed a specific column
df.rename(columns={"Foreign group":"Foreign"}, inplace=True) # inplace means we change the dataframe
````

### OUT OF PLACE

OUT OF PLACE will return a new dataframe, leaving the original dataframe unchanged. It is slower and uses more memory, but it is easier to understand and debug. It is recommended for production code.

````python
# we can also rename out of place
df = df.rename(columns={"Latvian group":"Latvian"}) # this will not change the original dataframe
# but since we overwrite the df variable we will not have access to the original dataframe
````

## Row indexing

In pandas we have two ways to index rows:

*  iloc - index by integer position (this is number)
*  loc - index by label (this could be number or string)

````python
# this is numeric row lookup no matter what the real index looks like
df.iloc[40:45] # just using our regular Python syntax for sequences square brackets
````

````python
# we could use also string index here they overlap
df.loc[40:45] # this is string based index so 45 is included here
# otherwise they are similar
# why?
# because our index is just a sequence of numbers
# generally you will want your index to be something more meaningful
# in this case it could be Year and Month combined
````

````python
df.index = df.Month # note that means our index could have multiple same values that is fine in Pandas
df.head()
````

````python
df.loc["Maijs"] # we utilize fact that our index is now a string of month names
````

## Creating new columns - feature engineering

In feature engineering we create new columns based on existing columns. This is done to improve model performance.

````python
# creating a new column
# df["all_groups"] = df["Foreign group"] + df["Latvian group"]
df["all_groups"] = df["Foreign"] + df["Latvian"] # so we create a new column by adding two existing columns
df.head()
````

## Create a numeric column from string column

````python
# lets create a numeric column form month names
# we will use a dictionary to map month names to numbers
month_map = {"Janvﾄ〉is":1, "Februﾄ〉is":2, "Marts":3, "Aprﾄｫlis":4, 
             "Maijs":5, "Jﾅｫnijs":6, "Jﾅｫlijs":7, "Augusts":8, 
             "Septembris":9, "Oktobris":10, "Novembris":11, "Decembris":12}
# we create a new column by mapping month names to number
df["Month_num"] = df.Month.map(month_map) # so we use map method to map month names to numbers
df.head()
````

````python
# let us now create a new column from year and Month_num
# we can always cast any data to string
# lets add 0 in front of single digit months
df["Month_num"] = df.Month_num.astype(str).str.zfill(2) # so we use string method zfill to add 0 in front of single digit months
df["Year_Month"] = "Year_" + df.Year.astype(str) + "_" + df.Month_num.astype(str) # so we use string concatenation
df.head()
````

````python
# Year_Month would make a good index
df.index = df.Year_Month
df.head()
````

````python
# let us group df by month and find some statistics
# we will use groupby method
# we will group by month and then we will aggregate by sum
# we will also reset index
# we will also sort by Pupils
grouped_df = df.groupby("Month").sum().reset_index().sort_values(by="Pupils", ascending=False)
grouped_df
````

````python
# we can reach a conclustion that most pupils come in May, September and October
# why is that?
# pupils come during school trips, in summer there are no school trips
# in summer pupils could be coming with their families
````

````python
# we can use other aggregation functions such as mean, count, min, max, etc
# you could even write your own aggregation functions
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.agg.html
````

````python
# we can also group by multiple columns
# we will group by year and month - in this case kind of redundant
# we will use following aggregation functions: sum, mean, count, min, max
# we will also reset index
# we will also sort by Pupils
grouped_df = (df.groupby(["Year", "Month"])
              .agg(["sum", "mean", "count", "min", "max"])
              .reset_index()
              .sort_values(by=("Pupils","sum"), ascending=False))
# i used parenthesis to break the code into multiple lines - little trick
# i had to use by=("Pupils","sum") because we have multiindex
grouped_df
````
