## Reading data from URL (CSV)

```python
# we can read data directlly from web url address
url = "https://data.gov.lv/dati/dataset/633b5005-fe7f-44bd-b17c-4995d359f9b0/resource/975fb5a2-1be4-4183-b404-b32afe60c01a/download/kopa_cesu_pils_apmekletaju_veidi_pa_gadiem_menesiem.csv"
df = pd.read_csv(url) # Cesis tourism data
df.head(3)# see  first 3 entrees
```

```python
# to fix encoding issues we can specify encoding
# to fix separator issues we can specify separator
df = pd.read_csv(url, sep=";",encoding="utf-8") # we download using ; as a separator for values (not the default comma)
df.head(3)# see  first 3 entrees
```

```python
df.columns
```

```python
df.columns = ["Month", "Foreign group", "Latvian group", "Family", "Couple", "Single", "Friends", "Pupils", "Year"]
df.head()
```

```python
df.shape
```

```python
df.Year.min(),df.Year.max() # this . syntax works with column names that are single word and do not conflict with dataframe methods
```

```python
df["Foreign group"].min(), df["Foreign group"].mean(), df["Foreign group"].max()
```

```python
df.describe() # numeric column stats
```

```python
df.describe(include="O") # so for Month only since that is the only string object column
```

```python
# this is numeric row lookup no matter what the real index looks like
df.iloc[40:45] # just using our regular Python syntax for sequences square brackets
```

```python
# we could use also string index here they overlap
df.loc[40:45] # this is string based index so 45 is included here
```

```python
df.index = df.Month
df.head()
```

```python
df.loc["Maijs"]
```

```python
df.iloc[2:5]  # 3rd to 5th row again numeric index even though our main index is now stringb
```

```python
df.loc[["Maijs","Jﾅｫnijs","Jﾅｫlijs"]]
```

```python
# we can filter by conditions inside
d2019 = df[df.Year == 2019]
d2019
```

```python
# filter by multiple conditions
big_pupils = df[(df.Year > 2015) & (df.Pupils > 1000)]  # notice the we use bit operator & instead of usual &
big_pupils
```

```python
df.Pupils > 1000 # this is Boolean series
```

```python
groups = df[["Latvian group", "Foreign group"]]  # so groups dataframe has 2 columsn in a different order from original df
groups.head()
```

```python
# creating a new column
df["all_groups"] = df["Foreign group"] + df["Latvian group"]
df.head()
```

```python
sorted_family = df.sort_values(by="Family")
sorted_family.head() 
```

```python
# i create a new data frame i could have also overwritend the old df  as well or use in place option
sorted_family = df.sort_values(by="Family", ascending=False ) # so from the top
sorted_family.head() 
```

```python
# sorting by multiple columns, with different ascending parameters
multiple_sort = df.sort_values(by=["Year", "all_groups"], ascending=[True,False]) # we have now tiebreaks
multiple_sort.head(15)
```
